```python
import pandas as pd
import numpy as np
import glob
import networkx as nx
import osmnx as ox
from shapely.geometry import Point
import matplotlib.pyplot as plt
import math
import transbigdata as tbd
```

```python
# 
#G = ox.graph_from_place('', network_type='drive')
#ox.save_graphml(G, 'shenzhen.graphml')
G = ox.load_graphml('shenzhen.graphml')
fig, ax = ox.plot_graph(G, figsize=(10, 10))
plt.show()
```

```python
entrance = pd.read_csv('entrance.csv')
entrance['node'] = ox.distance.nearest_nodes(G, entrance['longitude'], entrance['latitude'])
entrance.head(2)
```

```python
zone = pd.read_csv('zone-information.csv')
zone = zone[['TAZID', 'longitude', 'latitude']]
#
zone['node'] = ox.distance.nearest_nodes(G, zone['longitude'], zone['latitude'])
zone.head(2)
```

```python
def road_shortest(node):
    origin_node = node
    distance = []
    if entrance.empty:
        return np.nan
    else:
        for i in range(len(entrance)):
            destination_node = entrance['node'].iloc[i]
            try:
                route_length = nx.shortest_path_length(G, origin_node, destination_node, weight='length')
            except nx.NetworkXNoPath:
                route_length = np.nan
            distance.append(route_length)
        distance = [x for x in distance if isinstance(x, (int, float)) and not math.isnan(x)]
        return min(distance)/1000 if distance else None #km
```

```python
zone['distance'] = zone.apply(lambda zone:road_shortest(zone['node']), axis=1)
zone.head(2)
```

```python
zone1 = zone[~zone.isnull().any(axis=1)] 
zone2 = zone[zone.isnull().any(axis=1)]
print(zone1.shape)
print(zone2.shape)
```

```python
zone.to_csv('distance.csv',index=None)
```
